```python
# generating a sample signal with mix of frequencies
import numpy as np
def SinGenerator(f,sampling_rate,signal_length):
    s = list(np.linspace(0,2*np.pi*f,sampling_rate))
    s = np.sin(s*signal_length)
    return s

# first 3 seconds generated sin wave with freq 693,1023 and 1 second of silence and 2 seconds of sin wave with freq 697,1336 
sampling_rate = 4000
s_digit1 = SinGenerator(693,sampling_rate,3) + SinGenerator(1023,sampling_rate,3)
s_silence = [0]*sampling_rate*1
s_digit2 = SinGenerator(697,sampling_rate,2) + SinGenerator(1336,sampling_rate,2)

s = list(s_digit1) + list(s_silence) + list(s_digit2)
```

```python
from IPython.display import Audio
Audio(s, rate=sampling_rate)
```

###### Spectrogram:
 Spectrogram is a pictorial representation of specrum of frequencies which contains time in x-axis, frequency in y-axis and amplitute in the intensity of the colour. Spectrogram is generated by taking Fourier transform.

###### Generating spectrogram:
Creating a spectrogram using the FFT is a digital process. Digitally sampled data, in the time domain, is broken up into chunks, which usually overlap, and Fourier transformed to calculate the magnitude of the frequency spectrum for each chunk. Each chunk then corresponds to a vertical line in the image; a measurement of magnitude versus frequency for a specific moment in time (the midpoint of the chunk). These spectrums or time plots are then "laid side by side" to form the image or a three-dimensional surface or slightly overlapped in various ways.

##### Narrowband:
The longer the window length, the more precisely you can estimate the exact frequencies that are present within that window (and hence, the narrower the frequency bands have relatively longer window lengths). However, the whole window is analyzed as a chunk, and yields one column of values in the spectrogram, so longer windows means each column is "wider" in time, which means less precise information about when certain frequency components were loud or soft.


```python
import matplotlib.pyplot as plt


## used wider window size. we can notice that frequencies present are precisely seen (frequency resolution is high) 
## but time resolution is not good 


plt.specgram(s,Fs=sampling_rate,NFFT=1025)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar()
```

###### Wideband:
Shorter the window length then the time resolution is higher whereas the frequency resolution is not so good.

```python
plt.specgram(s,Fs=sampling_rate,NFFT=150)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar()
```
